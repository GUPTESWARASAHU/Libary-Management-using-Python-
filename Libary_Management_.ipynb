```python
# -----------------------------------------------------------
# Step 1: Import the os module
# -----------------------------------------------------------
import os

# -----------------------------------------------------------
# Step 2: Create empty library list
# -----------------------------------------------------------
library = []


# -----------------------------------------------------------
# Step 3: Function to add book
# -----------------------------------------------------------
def add_book(title, author, genre, year, status="unread"):
    book = {
        "title": title.strip(),
        "author": author.strip(),
        "genre": genre.strip(),
        "year": int(year),
        "status": status.strip().lower()
    }
    library.append(book)
    print(f"\nBook '{title}' added successfully!\n")


# -----------------------------------------------------------
# Step 4: List all books
# -----------------------------------------------------------
def list_books(sorted_by_year=False):
    if len(library) == 0:
        print("\nLibrary is empty.\n")
        return

    books_to_show = library

    if sorted_by_year:
        books_to_show = sorted(library, key=lambda x: x["year"])

    print("\nYour Library:\n")
    for i, book in enumerate(books_to_show, 1):
        print(f"{i}. {book['title']} by {book['author']} "
              f"[{book['genre']}, {book['year']}] – {book['status'].capitalize()}")
    print()


# -----------------------------------------------------------
# Step 5: Search by author
# -----------------------------------------------------------
def search_by_author(author):
    author = author.lower()
    results = [b for b in library if author in b["author"].lower()]
    return results


# -----------------------------------------------------------
# Step 6: Delete book by title
# -----------------------------------------------------------
def delete_book(title):
    global library
    title = title.lower()

    library = [b for b in library if b["title"].lower() != title]
    print(f"\nBook '{title}' removed if it existed.\n")


# -----------------------------------------------------------
# Step 7: Reading summary
# -----------------------------------------------------------
def reading_summary():
    if len(library) == 0:
        print("\nLibrary is empty.\n")
        return

    read_books = [b for b in library if b["status"] == "read"]
    unread_books = len(library) - len(read_books)

    read_percent = (len(read_books) / len(library)) * 100
    unread_percent = 100 - read_percent

    print("\nReading Summary:")
    print(f"Read: {len(read_books)} books ({read_percent:.2f}%)")
    print(f"Unread: {unread_books} books ({unread_percent:.2f}%)\n")


# -----------------------------------------------------------
# Step 8: Uppercase titles
# -----------------------------------------------------------
def uppercase_titles():
    if not library:
        print("\nLibrary is empty.\n")
        return

    print("\nBook Titles (UPPERCASE):")
    titles = map(lambda b: b["title"].upper(), library)
    for t in titles:
        print("-", t)
    print()


# -----------------------------------------------------------
# Step 9: Save library to file
# -----------------------------------------------------------
def save_to_file(filename):
    try:
        with open(filename, "w") as file:
            for b in library:
                line = f"{b['title']},{b['author']},{b['genre']},{b['year']},{b['status']}\n"
                file.write(line)

        print(f"\nSaved {len(library)} books to '{filename}'.\n")

    except Exception as e:
        print("Error saving file:", e)


# -----------------------------------------------------------
# Step 10: Load library from file
# -----------------------------------------------------------
def load_from_file(filename):
    if not os.path.exists(filename):
        print("\nFile not found. Starting with an empty library.\n")
        return

    try:
        with open(filename, "r") as file:
            for line in file:
                parts = line.strip().split(",")
                if len(parts) == 5:
                    title, author, genre, year, status = parts
                    add_book(title, author, genre, int(year), status)
        print(f"\nLibrary loaded from '{filename}'.\n")

    except Exception as e:
        print("\nThere was a problem reading the file.", e)


# -----------------------------------------------------------
# Step 11: Menu Function
# -----------------------------------------------------------
def menu():
    while True:
        print("========= PERSONAL LIBRARY MANAGER =========")
        print("1. Add Book")
        print("2. List All Books")
        print("3. List Books Sorted by Year")
        print("4. Search by Author")
        print("5. Delete Book by Title")
        print("6. Show Reading Summary")
        print("7. Show UPPERCASE Titles")
        print("8. Save Library")
        print("9. Load Library")
        print("10. Exit")
        print("=============================================")

        choice = input("Enter your choice (1–10): ")

        if choice == "1":
            title = input("Enter title: ")
            author = input("Enter author: ")
            genre = input("Enter genre: ")
            year = input("Enter year: ")
            status = input("Status (read/unread): ")

            add_book(title, author, genre, year, status)

        elif choice == "2":
            list_books(False)

        elif choice == "3":
            list_books(True)

        elif choice == "4":
            author = input("Enter author's name: ")
            results = search_by_author(author)

            if not results:
                print("\nNo books found.\n")
            else:
                print("\nSearch Results:")
                for b in results:
                    print(f"- {b['title']} by {b['author']}")
                print()

        elif choice == "5":
            title = input("Enter title to delete: ")
            delete_book(title)

        elif choice == "6":
            reading_summary()

        elif choice == "7":
            uppercase_titles()

        elif choice == "8":
            filename = input("Enter filename to save (e.g., books_data.txt): ")
            save_to_file(filename)

        elif choice == "9":
            filename = input("Enter filename to load: ")
            load_from_file(filename)

        elif choice == "10":
            print("\nExiting the program. Goodbye!\n")
       5     break

        else:
            print("\nInvalid option. Try again.\n")


# -----------------------------------------------------------
# Step 12: Run the app
# -----------------------------------------------------------
if __name__ == "__main__":
    menu()
```
